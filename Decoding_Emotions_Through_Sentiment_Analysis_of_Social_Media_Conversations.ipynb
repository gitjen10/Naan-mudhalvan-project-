```python
!pip install gradio emoji vaderSentiment

import gradio as gr
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
import emoji

# Emoji Sentiment Mapping
emoji_map = {
    "😊": "happy",
    "😢": "sad",
    "😡": "angry",
    "😂": "laugh",
    "😍": "love",
    "😞": "disappointed",
    "😭": "crying",
    "😤": "frustrated",
    "👍": "good",
    "👎": "bad",
    "🥳": "celebrating",
    "🤬": "furious",
    "😮": "surprised",
    "❤️": "love",
    "😐": "neutral"
}

# Emoji replacement function
def replace_emojis(text):
    for char in text:
        if char in emoji_map:
            text = text.replace(char, " " + emoji_map[char] + " ")
    return text

# Initialize VADER
analyzer = SentimentIntensityAnalyzer()

# Sentiment Classification
def classify_emotion(text):
    cleaned = replace_emojis(text)
    score = analyzer.polarity_scores(cleaned)
    compound = score['compound']
    if compound >= 0.05:
        return "😊 Positive"
    elif compound <= -0.05:
        return "😢 Negative"
    else:
        return "😐 Neutral"

# Reconstructed Interface
with gr.Blocks(theme=gr.themes.Soft()) as demo:
    gr.Markdown("# 🧠 Emotion Classifier with Emoji Support")
    gr.Markdown("Analyze emotions from text or emoji-rich messages. Powered by VADER.")

    with gr.Row():
        input_box = gr.Textbox(label="Enter your message", lines=3, placeholder="Type something like: I'm really excited 🥳")
        output_label = gr.Textbox(label="Predicted Emotion")

    submit_btn = gr.Button("🚀 Analyze Emotion")

    submit_btn.click(fn=classify_emotion, inputs=input_box, outputs=output_label)

demo.launch()
```
